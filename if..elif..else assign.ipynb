```python
#q3 
# fine amout based on speed
# input = 40
speed=int(input("Enter Speed of Vehicle :"))
if speed<=60:
    print("NO FINE")
elif speed>60 and speed==80:
    print("WARNING")
elif speed>80 and speed==100:
    print("Rs 500 FINE")
elif speed>100 and speed==120:
    print("Rs 1000 FINE")
else:
    print("CANCELATION OF LICESENCE")
```

```python
# q4
#input=
# number1 =1
#number2 =2
#RESULT = addition
number1=int(input("Enter First number:"))
number2=int(input("Enter Second Number:"))
RESULT=input("Give any arthematic operation (addition,substarction,multiplication or division):").lower()
if RESULT=="addition":
    print(number1 + number2)
elif RESULT=="substraction":
    print(number1 - number2)
elif RESULT=="multiplication":
    print(number1 * number2)
elif RESULT=="divison":
    print(number1/number2)
else:
    print("NO SUCH ARTHIMATIC OPERATIONS")
```
