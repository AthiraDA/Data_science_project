```python
import pandas as pd
salaries=pd.read_csv("Salaries Complete.csv")
salaries
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#count plot- only one axis and uses categorical value
plt.figure(figsize=(8,4))
sns.countplot(x='Degree',data=salaries,color='blue')
plt.title('COUNT BY DEGREE')
plt.xlabel('Degree')
plt.ylabel('count')
plt.show()
#INFERENCE : From the plot we infer that the no of students pursued diploma is more than Bsc and Msc
```

```python
#BAR PLOT - two axis x= categorical,y=integer
plt.Figure(figsize=(10,4))
sns.barplot(x='Degree',y='Experience',data=salaries,color='red')
plt.title('STUDENTS PURSUED DEGREE AND HAS EXPIERENCE')
plt.show()
#INFERENCE: from the plot we infer that the stuents pursued diploma has more years of expiernce compared to others
```

```python
#histogram- only one axis and uses only integer
plt.figure(figsize=(7,4))
sns.histplot(salaries['Allowance'],bins=10,kde=True,color='green')
plt.title('ALLOWANCE PROVIDED')
plt.show()
#INFERENCE- THERE IS A BELL SHAPED CURVE WHICH EXTENDS TOWARDS RIGHT MEANS + SKWENESS AND THE MEAN IS LESS THAN THE MEDIAN
```

```python
#pie chart - only categorical
marital_count=salaries['Marital Status'].value_counts()
plt.figure(figsize=(10,4))
plt.pie(marital_count,labels=['married','single'],autopct='%1.1f%%')
plt.title('Marital Status')
plt.show()
#INFERENCE FROM THIS WE INFER THAT THE % OF MARRIED COMPARED TO UNMARRIED
```

```python
#BOXPLOT-only one axis and integer
plt.figure(figsize=(8,2))
sns.boxplot(x='Family',data=salaries,palette='Set1')
plt.title('salary allocated to family')
plt.show()
#INFERENCE: therefore the salary allocated for family is mostly between 0 to 95
```

```python
#scatterplot= two axis only integer
plt.Figure(figsize=(8,4))
sns.scatterplot(x='Experience',y='Transportation',hue='Marital Status',style='Gender',data=salaries,palette='rainbow')
plt.title('expirence vs transportation by marital status and gender')
plt.show()
#inference- we infer that there is no correlation 
```

```python
#heatmap- integer
numerical=salaries[['Experience','Dependents','Housing','Transportation','Basic','Allowance','Family']].corr()
plt.figure(figsize=(10,4))
sns.heatmap(numerical,annot=True,cmap='rainbow',center=2)
plt.show()
```

```python
#violin plot- categorical and integer
plt.Figure(figsize=(8,4))
sns.violinplot(x="Marital Status",y='Dependents',hue='Gender',split=False,data=salaries,palette='rainbow')
plt.show()
```
