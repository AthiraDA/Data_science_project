```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
hotel_booking['country']=le.fit_transform(hotel_booking['country'])
hotel_booking['hotel']=le.fit_transform(hotel_booking['hotel'])
hotel_booking['arrival_date_month']=le.fit_transform(hotel_booking['arrival_date_month'])
hotel_booking['meal']=le.fit_transform(hotel_booking['meal'])
hotel_booking['market_segment']=le.fit_transform(hotel_booking['market_segment'])
hotel_booking['distribution_channel']=le.fit_transform(hotel_booking['distribution_channel'])
hotel_booking['reserved_room_type']=le.fit_transform(hotel_booking['reserved_room_type'])
hotel_booking['assigned_room_type']=le.fit_transform(hotel_booking['assigned_room_type'])
hotel_booking['deposit_type']=le.fit_transform(hotel_booking['deposit_type'])
hotel_booking['reservation_status']=le.fit_transform(hotel_booking['reservation_status'])
hotel_booking['reservation_status_date']=le.fit_transform(hotel_booking['reservation_status_date'])
hotel_booking['city']=le.fit_transform(hotel_booking['city'])
hotel_booking['city']=le.fit_transform(hotel_booking['city'])
hotel_booking['customer_type']=le.fit_transform(hotel_booking['customer_type'])
```

```python
hotel_booking['country']=hotel_booking['country'].fillna(hotel_booking['country'].median())
hotel_booking['children']=hotel_booking['children'].fillna(hotel_booking['children'].median())
```

```python

h_b=hotel_booking.drop(['company','agent'],axis=1)
```

```python
h_b.isnull().sum()
```

```python
hotel_booking.info()
```

```python
h_b['is_canceled'].value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler
x=h_b.drop(['is_canceled'],axis=1)
y=h_b['is_canceled']
smot=RandomUnderSampler(sampling_strategy='majority')
input,output=smot.fit_resample(x,y)
```

```python
print(input.shape)
print(output.shape)
```

```python
output.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
