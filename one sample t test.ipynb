```python
#labelEncoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
student_exam['student_id']=le.fit_transform(student_exam['student_id'])
```

```python
student_exam.shape
```

```python
# 2 sample t- test
import numpy as np 
sample_size=75
x=np.random.choice(student_exam['hours_studied'],sample_size)
y=np.random.choice(student_exam['exam_score'],sample_size)
print(x)
print(y)
```

```python
from scipy import stats
t_stat,p_value=stats.ttest_ind(x,y)
print("t_statistics :",round(t_stat,2))
print("p-value :",round(p_value,4))
# decision 
alpha=0.05
if p_value < alpha:
    print("Reject H0 and there is significant difference between hours studied and exam score")
else:
    print("Accept H0 and Reject H1,there is no  significant difference between hours studied and exam score")
```

```python
# 1 sample t test
sample_size=75
sample_population=np.random.choice(student_exam['sleep_hours'],sample_size)
```

```python
sample_population
```

```python
population_mean=np.mean(student_exam['sleep_hours'])
population_mean
```

```python
# 1 sample t test
# sample population= 75
# population mean= 33.955
t_stat,p_value=stats.ttest_1samp(sample_population,population_mean)
print("t_statistics :",round(t_stat,4))
print("p_value :",round(p_value,4))
# decision 
alpha=0.05
if p_value < alpha:
    print("Reject H0 and there is significant difference in sleep hours")
else:
    print("Accept H0 and Reject H1,there is no significant difference in sleep hours")
```
