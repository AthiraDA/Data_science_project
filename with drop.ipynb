```python
house_price.drop(columns=['Date'],inplace=True,errors='ignore')
```

```python
house_price.drop(columns=['bathrooms'],inplace=True)
house_price.drop(columns=['yr_built'],inplace=True)
house_price.drop(columns=['city'],inplace=True)
house_price.drop(columns=['street'],inplace=True)
house_price.drop(columns=['country'],inplace=True)
house_price.drop(columns=['statezip'],inplace=True)
```

```python
house_price
```

```python
x=house_price.drop(['price'],axis=1)
y=house_price['price']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
z=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from  sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error 
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
lr=LinearRegression()
lr.fit(x_test,y_test)
y_pred=lr.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
from  sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
